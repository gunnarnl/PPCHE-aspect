# Cleaning up tregex outputs

Individual matches are separated by filename prefixed with a `#`. We can just split the documents there.

```python
import re
import pandas as pd
```

```python
def get_tregex_matches(text):
    return re.split("# ", text)
```

```python
def make_matches(filename):
    with open(filename) as f:
        text = f.read()
        matches = get_tregex_matches(text)
        return matches[1:] #first list element always blank
```

```python
def split_matches(matches, min_items): #min_items is the minimum number of elements in a match for a given file
    split_matches = []
    for m in matches:
        items = re.split('\n', m)[:-1]
        if len(items) < min_items:
            print("Too few elements!")
            break
        if len(items) == min_items:
            items.extend([None, None])
        if len(items) == min_items+1:
            if re.match('\(MD', items[min_items]):
                items.append(items[min_items])
                items[min_items] = None
            else:
                items.append(None)
#                 items.append(items[min_items])
#                 items[min_items] = None
        split_matches.append(items)
    return split_matches
```

```python
def matches_to_pd(matches, columns):
    df = pd.DataFrame(matches, columns=columns)
    return df
```

```python
def make_dfs(treg_dict):
    dfs = dict()
    for k in treg_dict:
        matches = split_matches(make_matches(treg_dict[k]['filename']), len(treg_dict[k]['cols'])-2)
        dfs[k] = matches_to_pd(matches, treg_dict[k]['cols'])
        if (k=='vbd') | (k=='vbp') | (k=='prf') | (k=='md'):
            dfs[k]['be'] = None
        if (k=='vbd') | (k=='vbp') | (k=='prog') | (k=='md'):
            dfs[k]['hv'] = None
        dfs[k]['type'] = k
    return dfs
```

## Fill out dataframes

```python
def clean_id(id_raw):
    clean_id = []
    id_text = id_raw[4:-1]
    txt_pgtok = re.split(',',id_text,1)
    pg_tok = re.split('\.', txt_pgtok[1],1)
    clean_id.append(txt_pgtok[0])
    clean_id.extend(pg_tok)
    return clean_id
```

```python
def get_aux_pos(aux_raw):
    if aux_raw==None:
        return None
    if aux_raw[3] == ' ':
        return aux_raw[1:3]
    else:
        return aux_raw[1:4]
```

```python
def set_tense(df):
    df['tense'] = '?'
    df['tense'].loc[df['be_pos'].str.contains('P') | df['hv_pos'].str.contains('P') | (df['type']=='vbp')] = 'P'
    df['tense'].loc[df['be_pos'].str.contains('D') | df['hv_pos'].str.contains('D') | (df['type']=='vbd')] = 'D'
    df['tense'].loc[df['md'].notna()] = 'M'
    return
```

```python
# Load wordcount document
wc_ppcme2 = pd.read_csv('/home/gunnar/Projects/corpora/PPCHE/PPCHE/PENN-CORPORA/PPCME2-RELEASE-4/WORDCOUNT-PPCME2', sep='\s', header=None, skiprows=1)
wc_ppcme2.columns = ['id_txt', 'date', 'genre', 'wordcount']

# There are no texts in the txt_id field not also in this document.
wc_ppcmbe2 = pd.read_csv('/home/gunnar/Projects/corpora/PPCHE/PPCHE/PENN-CORPORA/PPCMBE2-RELEASE-1/WORDCOUNT-PPCMBE2', sep='\s', header=None)
wc_ppcmbe2.columns = ['id_txt', 'date', 'genre', 'wordcount']

#for some reason this one has a header
wc_ppceme = pd.read_csv('/home/gunnar/Projects/corpora/PPCHE/PPCHE/PENN-CORPORA/PPCEME-RELEASE-3/WORDCOUNT-PPCEME', sep='\s', header=None, skiprows=1)
wc_ppceme.columns = ['id_txt', 'date', 'genre', 'wordcount']

# fix years that are listed as ranges, with 'c', or whatever. Just take the first sequence of 4 numerals.
def fix_date(date):
    if len(date) > 4:
        mobj = re.search('[0-9]{4}', date)
        return mobj.group()
    else:
        return date

# We can't use datetime because it doesn't support older dates...
#wc_ppceme['date'] = pd.to_datetime(wc_ppceme['date'].apply(fix_date)) #datetime dtype

wc_ppcme2['date'] = wc_ppcme2['date'].apply(fix_date)
wc_ppcmbe2['date'] = wc_ppcmbe2['date'].apply(fix_date)
wc_ppceme['date'] = wc_ppceme['date'].apply(fix_date)
wc_ppcme2.head()
```

```python
def set_docinfo(df, wc):
    return pd.merge(df, wc, how='inner', on='id_txt')
```

```python
def fill_dataframe(dfo, wc):
    df = dfo
    df[['id_txt', 'id_pg', 'id_tok', 'be_pos', 'hv_pos']] = None
    id_df = pd.DataFrame(df['id_raw'].apply(clean_id).tolist(), columns=['id_txt', 'id_pg', 'id_tok'])
    df[['id_txt', 'id_pg', 'id_tok']] = id_df
    df['id_txt'] = df['id_txt'].str.lower()
    df['be_pos'] = df['be'].apply(get_aux_pos)
    df['hv_pos'] = df['hv'].apply(get_aux_pos)
    set_tense(df)
    df = set_docinfo(df, wc)
    return df
```

```python
def build_subcorp_df(df_dict, wc):
    dfs = []
    for k in df_dict:
        dfs.append(fill_dataframe(df_dict[k], wc))
    return pd.concat(dfs)
```

```python
ppcme2 = build_subcorp_df(ppcme2_dfs, wc_ppcme2)
ppceme = build_subcorp_df(ppceme_dfs, wc_ppceme)
ppcmbe2 = build_subcorp_df(ppcmbe2_dfs, wc_ppcmbe2)
```
